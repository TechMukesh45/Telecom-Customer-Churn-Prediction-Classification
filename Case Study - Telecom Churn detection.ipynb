# hypothesis tesing 

```python
from scipy.stats import ttest_ind

# Subset the data for customers who churned and those who did not
churned = df[df['CHURN STATUS'] == 'Yes']
not_churned = df[df['CHURN STATUS'] == 'No']

# Define null and alternative hypotheses
null_hypothesis = "There is no significant difference in payment behavior (PAY_1 to PAY_6) between customers who churned and those who did not."
alternative_hypothesis = "There is a significant difference in payment behavior (PAY_1 to PAY_6) between customers who churned and those who did not."

# Perform t-test for PAY variables between churned and not churned customers
t_statistics = {}
p_values = {}
for i in range(1, 7):
    t_statistic, p_value = ttest_ind(churned[f'PAY_{i}'], not_churned[f'PAY_{i}'], equal_var=False)
    t_statistics[f'PAY_{i}'] = t_statistic
    p_values[f'PAY_{i}'] = p_value

# Print the hypotheses
print("Null Hypothesis:", null_hypothesis)
print("Alternative Hypothesis:", alternative_hypothesis)

# Print the results for each PAY variable
print("\nResults:")
for i in range(1, 7):
    print(f"PAY_{i}:")
    print("t-statistic:", t_statistics[f'PAY_{i}'])
    print("p-value:", p_values[f'PAY_{i}'])
    alpha = 0.05  # significance level
    if p_values[f'PAY_{i}'] < alpha:
        print("Conclusion: Reject the null hypothesis.")
        print("There is sufficient evidence to suggest that", alternative_hypothesis)
    else:
        print("Conclusion: Fail to reject the null hypothesis.")
        print("There is not enough evidence to support", alternative_hypothesis)
    print()  # Print an empty line for readability
```

**Conclusion -** 
Payment Behavior plays a significant role in determining customer churn.

# 4.5 Hyperparamter tuning 

based on obervations model selected for hyperparameter tuning is random forest 

```python


# Define a smaller parameter grid for Random Forest
rf_param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [None, 10],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'bootstrap': [True, False]
}

# Initialize Random Forest Classifier
random_forest_classifier = RandomForestClassifier(random_state=42)

# Initialize GridSearchCV for Random Forest
rf_grid_search = GridSearchCV(random_forest_classifier, rf_param_grid, cv=5, scoring='accuracy', n_jobs=-1)

# Fit the model
rf_grid_search.fit(X_train, y_train)

# Print the best parameters and the corresponding accuracy
print("Random Forest - Best Parameters:", rf_grid_search.best_params_)
print("Random Forest - Best Accuracy:", rf_grid_search.best_score_)
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Get predicted probabilities for the positive class
y_pred_rf_prob = rf_grid_search.predict_proba(X_test)[:, 1]

# Compute ROC curve and AUC
fpr_rf, tpr_rf, thresholds_rf = roc_curve(y_test, y_pred_rf_prob)
roc_auc_rf = auc(fpr_rf, tpr_rf)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr_rf, tpr_rf, color='darkorange', lw=2, label='ROC curve (area = {:.2f})'.format(roc_auc_rf))
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Random Forest Classifier')
plt.legend(loc='lower right')
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV

# Define a smaller parameter grid for Gradient Boosting
gb_param_grid = {
    'n_estimators': [50, 100],
    'learning_rate': [0.01, 0.1],
    'max_depth': [3, 5],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

# Initialize Gradient Boosting Classifier
gradient_boosting_classifier = GradientBoostingClassifier(random_state=42)

# Initialize GridSearchCV for Gradient Boosting
gb_grid_search = GridSearchCV(gradient_boosting_classifier, gb_param_grid, cv=5, scoring='accuracy', n_jobs=-1)

# Fit the model
gb_grid_search.fit(X_train, y_train)

# Print the best parameters and the corresponding accuracy
print("Gradient Boosting - Best Parameters:", gb_grid_search.best_params_)
print("Gradient Boosting - Best Accuracy:", gb_grid_search.best_score_)
```

```python
# Get predicted probabilities for the positive class
y_pred_gb_prob = gb_grid_search.predict_proba(X_test)[:, 1]

# Compute ROC curve and AUC
fpr_gb, tpr_gb, thresholds_gb = roc_curve(y_test, y_pred_gb_prob)
roc_auc_gb = auc(fpr_gb, tpr_gb)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr_gb, tpr_gb, color='darkorange', lw=2, label='ROC curve (area = {:.2f})'.format(roc_auc_gb))
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Gradient Boosting Classifier')
plt.legend(loc='lower right')
plt.show()
```
